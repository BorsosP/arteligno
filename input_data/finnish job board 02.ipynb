```python
url = 'https://paikat.te-palvelut.fi/tpt/?professions=251%2C252%2C351%2C352&announced=0&leasing=0&english=false&sort=1'

# create a new Firefox session
driver = webdriver.Chrome()
driver.implicitly_wait(30)
driver.get(url)
        
```

```python
soup = BeautifulSoup(driver.page_source, 'lxml')
        
job_title_results = soup.find_all('div', attrs={'class':'col-xs-12 list-group-item'}) 
```

```python
pos_all = []
for result in job_title_results:
    position = result.find('h5').text
    job_description_url = result.find('a').get('href')
    company = result.find('p').text
    pos_all.append([position, job_description_url, company])
```

```python
role_dataframe_page = pd.DataFrame(pos_all, columns=['position', 'url', 'other'])
```

```python
role_dataframe_page['other'] = role_dataframe_page['other'].apply(lambda text: re.sub('\n','', text))
splitted = role_dataframe_page['other'].str.split('-', expand=True)
splitted.rename(columns={0 : 'company', 1 : 'location'}, inplace=True)
role_dataframe_page = pd.merge(role_dataframe_page, splitted, left_index=True, right_index=True)
role_dataframe_page.fillna('mv-9', inplace=True)
role_dataframe_page = role_dataframe_page.applymap(lambda x: x.strip())
role_dataframe_page = role_dataframe_page[['position', 'url', 'company', 'location']]
```

```python
def description_extractor(url):
    try:
        print(url)
        driver.get('https://paikat.te-palvelut.fi' + url)
        
        print('wait 5 seconds')
        time.sleep(5)
        soup = BeautifulSoup(driver.page_source, 'lxml')

        job_desc_results = soup.find('p', attrs={'class':'detailText'}).text

        job_desc_results = re.sub('\n', '', job_desc_results.strip())
        print('wait 5 sec')
        time.sleep(5)
        return job_desc_results
    except Exception as e:
        logging.exception("message")
        print("Error detected and logged")
```

```python
role_dataframe_all['description'] = role_dataframe_all['url'].apply(description_extractor)
```

```python
role_dataframe_all['url'] = role_dataframe_all.apply(lambda df: 'https://paikat.te-palvelut.fi' + df['url'], axis=1)
```
