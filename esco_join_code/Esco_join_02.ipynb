```python
import pandas as pd
import numpy as np
import sqlite3
import glob
```

```python
#purpose of this function is to read in the multiple language esco files. Same format but different language. 
#At this point it's written to excel so the language codes could be assigned manually

def file_reader(folder):
    path =r'C:\Users\cameg\python_projects\esco_join_code' + '\\' + folder  # use your path
    allFiles = glob.glob(path + "/*.csv")
    complete_data = pd.DataFrame()
    list_of_files = []
    for file_ in allFiles:
        df = pd.read_csv(file_,index_col=None, header=0)
        list_of_files.append(df)
    complete_data = pd.concat(list_of_files)
    complete_data.to_excel('merged_data_' + folder + '.xlsx')
```

```python
#list of folders where the esco files are copies
folder_list = ['ISCO_GROUP', 'OCCUPATIONS', 'SKILLS', 'SKILLS_GROUPS']
```

```python
#run file reader function to all folders
for folder in folder_list:
    file_reader(folder)
```

```python
#After assigning the language codes manually read in all files

occupations = pd.read_excel('merged_data_OCCUPATIONS.xlsx')
skill_groups = pd.read_excel('merged_data_SKILLS_GROUPS.xlsx')
skills = pd.read_excel('merged_data_SKILLS.xlsx')
isco_groups = pd.read_excel('merged_data_ISCO_GROUP.xlsx')
broader_rel_occ = pd.read_csv('broaderRelationsOccPillar.csv')
broader_rel_skill = pd.read_csv('broaderRelationsSkillPillar.csv')
occ_to_skill = pd.read_csv('occupationSkillRelations.csv')
```

```python
#read in the IT categories which for I filter the dataset later on
isco_checked = pd.read_excel('isco_labels_checked.xlsx')
```

```python
#join occupations and the ISCO categories
isco_groups.rename(columns={'code' : 'iscoGroup'}, inplace=True)
occupations_w_isco = pd.merge(occupations, isco_groups, on=['iscoGroup', 'language'], how='left')
```

```python
#Join columns which will enable to full join for skills to ISCO and Occupations 
occupations_w_isco.rename(columns={'conceptUri_x' : 'occupationUri'}, inplace=True)
occupations_w_isco_w_skill = pd.merge(occupations_w_isco, occ_to_skill, on=['occupationUri'], how='left' )
```

```python
#Join all skill related field
skills.rename(columns={'conceptUri' : 'skillUri'}, inplace=True)
occupations_w_isco_w_skill = pd.merge(occupations_w_isco_w_skill, skills, on=['skillUri', 'language'], how='left')
```

```python
#formatting
occupations_w_isco_w_skill = occupations_w_isco_w_skill.astype(str)
occupations_w_isco_w_skill = occupations_w_isco_w_skill.applymap(lambda x: str(x.replace('\n', ', ')))
```

```python
#Rename columns where there was a naming conflict in the previous joins

occupations_w_isco_w_skill.rename(columns={'conceptType_x' : 'conceptType_occ',
                                          'preferredLabel_x' : 'preferred_occupation_label',
                                          'altLabels_x' : 'alt_occupation_label',
                                          'description_x' : 'occupation_description',
                                           'conceptType_y' : 'conceptType_isco',
                                          'conceptUri_y' : 'isco_uri',
                                          'preferredLabel_y' : 'preferred_isco_label',
                                          'altLabels_y' : 'alt_isco_label',
                                          'description_y' : 'isco_description',
                                          'relationType' : 'skill_relation',
                                          'preferredLabel' : 'preferred_skill_label',
                                          'altLabels' : 'alt_skill_labels',
                                          'description' : 'skill_description'}, inplace=True)
```

```python
#Write merged esco dataset to csv
occupations_w_isco_w_skill.to_csv('esco_joined_data_multi_lang_2018_04_28.csv', sep='^', encoding='utf-8')
```
